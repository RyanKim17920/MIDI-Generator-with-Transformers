```python
all_data = np.array([])
for i in range(len(short_data)):
    all_data = np.append(all_data, short_data[i])
all_data = all_data.reshape(-1, 11)
```

```python
tokenizer = Tokenizer()
tokenizer.fit(all_data)
```

```python
print(len(tokenizer.num_to_values))
print(len(tokenizer.num_to_times))
print(len(tokenizer.num_to_instruments))
```

```python
tokenized_short_data = tokenizer.tokenize_all(short_data)
```
