```python
import numpy as np
import pickle
np.set_printoptions(threshold=np.inf)
np.set_printoptions(linewidth=np.inf)
from tokenizer import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
with open('all_data.pickle', 'rb') as f:
    all_data = pickle.load(f)
```

```python
for i in range(len(all_data)):
    if len(all_data[i])>2**13:
        all_data[i] = all_data[i][:2**13]
```

```python
full_data = np.array([])
for i in range(len(all_data)):
    full_data = np.append(full_data, all_data[i])
full_data = full_data.reshape((-1,11))
```

```python
tokenizer.fit(full_data)
```

```python
with open('mid_tokenizer.pickle', 'wb') as f:
    pickle.dump(tokenizer, f)
```

```python
tokenized_list = tokenizer.tokenize_all(all_data)
```

```python
with open('mid_tokenized_list.pickle', 'wb') as f:
    pickle.dump(tokenized_list, f)
```

```python
with open('mid_tokenized_list.pickle', 'rb') as f:
    tokenized_list = pickle.load(f)
```
