```python
import numpy as np
import pickle
np.set_printoptions(threshold=np.inf)
np.set_printoptions(linewidth=np.inf)
```

```python
all_data_o = np.load('all_unique_data.npy', allow_pickle=True)
```

```python
all_data_o[0]
```

```python
all_data = all_data_o.reshape((-1, 15))
#change dtype of all_data to int64

all_data = all_data.astype(np.int64)
```

```python
# sort all_data so that the value are in order in this order: index 5, index 4

all_data_2 = all_data[np.lexsort((-all_data[:,4], -all_data[:,5]))]
```

```python
all_data_2 = all_data_2.reshape((-1, 15))
```

```python
all_data_2[0:10000]
```

```python
# pickle values up to index 11 (values), pickle index 12 (time), pickle values from index 14 to end (instruments)
```

```python
num_to_values = {}
values_to_num = {}
time_to_num = {}
num_to_times = {}
num_to_instruments = {}
instruments_to_num = {}
```

```python
values_count = 0
time_count = 0
instruments_count = 0
```

```python
values_array = np.unique(all_data_2[:,0:12], axis=0)
```

```python
values_array[0]
```

```python
values_array = values_array[np.lexsort((-values_array[:,4], -values_array[:,5]))]
```

```python
#swap index 0 and 1 of values_array
values_array[0][5]=0
values_array[1][5]=1
# so start of time is index 0, end of time index 1
```

```python
values_array[0:1000]
```

```python
for i in range(len(values_array)):
    values_to_num[str(values_array[i])] = values_count
    num_to_values[values_count] = values_array[i]
    values_count += 1
```

```python
print(len(values_array))
print(len(values_to_num))
print(len(num_to_values))
```

```python
#pickle the two dictionaries

with open('values_to_num.pickle', 'wb') as handle:
    pickle.dump(values_to_num, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('num_to_values.pickle', 'wb') as handle:
    pickle.dump(num_to_values, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
time_array = np.sort(np.unique(all_data_2[:,12], axis=0))
```

```python
time_array
```

```python
for i in range(len(time_array)):
    time_to_num[str(time_array[i])] = time_count
    num_to_times[time_count] = time_array[i]
    time_count += 1
```

```python
print(len(time_array))
print(len(time_to_num))
print(len(num_to_times))
```

```python
#pickle the two dictionaries

with open('time_to_num.pickle', 'wb') as handle:
    pickle.dump(time_to_num, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('num_to_times.pickle', 'wb') as handle:
    pickle.dump(num_to_times, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
instruments_array = np.unique(all_data_2[:,13:], axis=0)
```

```python
instruments_array
```

```python
for i in range(len(instruments_array)):
    instruments_to_num[str(instruments_array[i])] = instruments_count
    num_to_instruments[instruments_count] = instruments_array[i]
    instruments_count += 1
```

```python
print(len(instruments_array))
print(len(instruments_to_num))
print(len(num_to_instruments))
```

```python
#pickle the two dictionaries

with open('instruments_to_num.pickle', 'wb') as handle:
    pickle.dump(instruments_to_num, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('num_to_instruments.pickle', 'wb') as handle:
    pickle.dump(num_to_instruments, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
