```python
import os
import numpy as np
from MIDI_data_extractor import MIDI_data_extractor
import pickle   
import matplotlib.pyplot as plt
```

```python
music_len_to_pairing = {}
```

```python
def count_files_in_folder(path):
    count = 0
    extra = 0
    for root, dirs, files in os.walk(path):
        for file in files:
            if file.endswith(".mid"):
                count += 1
            else:
                extra += 1
                print(file)

    return count, extra

print(count_files_in_folder(r"C:\Users\ilove\CODING\PYStuff\MusicNet\Midi2Numpy\MIDI-Generator-with-Transformers\Bach MIDIs"))
```

```python
def extract_all_to_nums(input_path):
    # go to all MIDI files in folder and extract data from path, go to all subfolders and do the same
    # return numpy array of all data
    # Note: MIDI_data_extractor takes in input_path of MIDI file
    os.chdir(r"C:\Users\ilove\CODING\PYStuff\MusicNet\Midi2Numpy\MIDI-Generator-with-Transformers")
    os.chdir(input_path)
    cur_data = []
    try:
        for root, dirs, files in os.walk(input_path):
            #print(dirs,files)
            for file in files:
                if file.endswith(".mid") or file.endswith(".MID"):
                    data = MIDI_data_extractor(os.path.join(root, file), 0, True)
                    #check if any of data's indexes -4 is negative or greater than 10000
                    #if so, print the file name and index
                    if np.any(data[:, -4] < -1):
                        print("Negative????")
                        print(file)
                        print(root)
                    cur_data.append(data)
                    music_len_to_pairing[file] = len(data)
    except KeyboardInterrupt:
        return cur_data
    return cur_data
all_data = (extract_all_to_nums(r"C:\Users\ilove\CODING\PYStuff\MusicNet\Midi2Numpy\MIDI-Generator-with-Transformers\Bach MIDIs"))
```

```python
with open('music_len_to_pairing.pickle', 'wb') as f:
    pickle.dump(music_len_to_pairing, f)
```
